```python
import plotly.express as px
import pandas as pd

# 1. Criando dados de exemplo (Simulando uma análise de vendas)
data = {
    'Investimento_Marketing': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Retorno_Vendas': [110, 190, 310, 380, 550, 590, 850, 820, 1050, 1100],
    'Mes': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out'],
    'Categoria': ['Digital', 'TV', 'Digital', 'Digital', 'TV', 'Digital', 'TV', 'TV', 'Digital', 'Digital']
}
df = pd.DataFrame(data)

# 2. Criando o gráfico com Plotly Express
fig = px.scatter(df, 
                 x="Investimento_Marketing", 
                 y="Retorno_Vendas", 
                 color="Categoria",    # Cores diferentes por categoria
                 text="Mes",           # Mostra o mês ao lado do ponto
                 title="Investigação: Investimento vs Retorno",
                 labels={"Investimento_Marketing": "Investimento (R$)", "Retorno_Vendas": "Vendas (R$)"},
                 trendline="ols")      # Adiciona uma linha de tendência (previsão simples)

# 3. Exibindo o gráfico
fig.show()
```
